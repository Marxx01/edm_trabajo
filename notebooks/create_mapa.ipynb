```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster

# Cargar CSV
df = pd.read_csv("../data/csv/barris-policials.csv", sep = ";")

# Separar latitud y longitud
df[['LATITUD', 'LONGITUD']] = df['geo_point_2d'].str.split(',', expand=True).astype(float)

# Crear mapa centrado
mapa = folium.Map(location=[df['LATITUD'].mean(), df['LONGITUD'].mean()], zoom_start=13)

# Crear el cluster de marcadores
marker_cluster = MarkerCluster().add_to(mapa)

# Añadir marcadores al cluster
for _, row in df.iterrows():
    folium.Marker(
        location=[row['LATITUD'], row['LONGITUD']],
        popup=f"Barrio: {row.get('Nom_Barri', 'N/D')}",  # puedes añadir más info
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(marker_cluster)

mapa
```
